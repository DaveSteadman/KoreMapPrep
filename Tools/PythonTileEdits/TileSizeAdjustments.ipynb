```python
from PIL import Image

def is_blue(pixel, factor=1.1):
    r, g, b, *_ = pixel
    return b > (r * factor) and b > (g * factor)

def blue_percentage(image_path, threshold):
    with Image.open(image_path) as img:
        pixels = img.load()
        total_pixels = img.width * img.height
        blue_pixels = sum(1 for x in range(img.width) for y in range(img.height) if is_blue(pixels[x, y]))
        blue_percentage = blue_pixels / total_pixels
        return blue_percentage
```

```python

from PIL import Image

def nearest_power_of_two(x, min_size=128, max_size=2048):
    # Start with the smallest power of two greater than or equal to min_size
    size = min_size
    while size < x:
        size *= 2
    return min(max_size, size)

def nearest_lower_power_of_two(x):
    power = 128  # Start with the smallest power of two you want to support
    while power * 2 <= x:
        power *= 2
    return power

def resize_to_nearest_power_of_two(image_path, output_path):
    # Open the image
    with Image.open(image_path) as img:
        # Get the original dimensions
        original_width, original_height = img.size

        # Calculate the nearest powers of two for both dimensions
        new_width = nearest_power_of_two(original_width)
        new_height = nearest_power_of_two(original_height)

        # Resize the image
        resized_img = img.resize((new_width, new_height), Image.LANCZOS)

        # Save the resized image
        resized_img.save(output_path, format='PNG')
        print(f"Image resized from {original_width}x{original_height} to {new_width}x{new_height} and saved to {output_path}")

```

```python
import os
def resize_pngs_in_directory(directory):
    # Traverse the directory hierarchy
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith('.png'):
                file_path = os.path.join(root, file)
                print(f"Processing {file_path}")
                resize_to_nearest_power_of_two(file_path, file_path)
```

```python
from PIL import Image
import os

def is_blue(pixel, factor=1.1):
    r, g, b, *_ = pixel
    return b > (r * factor) and b > (g * factor)

def blue_percentage(image_path, threshold):
    with Image.open(image_path) as img:
        pixels = img.load()
        total_pixels = img.width * img.height
        blue_pixels = sum(1 for x in range(img.width) for y in range(img.height) if is_blue(pixels[x, y]))
        blue_percentage = blue_pixels / total_pixels
        return blue_percentage

def traverse_and_check(directory, threshold, perform_resize):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".png"):
                file_path = os.path.join(root, file)
                blue_pct = blue_percentage(file_path, threshold)  # get the percentage
                blue_check = blue_pct >= threshold
                action = f"{blue_pct * 100:.2f}% - Resized to 256x256" if (perform_resize and blue_check) else f"{blue_pct * 100:.2f}% - Will resize to 512x512" if blue_check else f"{blue_pct * 100:.2f}% - NOT BLUE"
                print(f"{file_path}: {action}")
                if perform_resize and blue_check:
                    with Image.open(file_path) as img:
                        img_resized = img.resize((256, 256), Image.LANCZOS)
                        img_resized.save(file_path)
```

```python
from PIL import Image
import os

def resize_if_sea(image_path, threshold=0.95):
    blue_percent = blue_percentage(image_path, threshold)

    if blue_percent >= threshold:
        with Image.open(image_path) as img:
            # Calculate the new dimensions
            new_width  = 256
            new_height = 256

            # Resize and overwrite the image if it's smaller
            if new_width < img.width or new_height < img.height:
                resized_img = img.resize((new_width, new_height), Image.LANCZOS)
                resized_img.save(image_path, format='PNG')
                print(f"Image at {image_path} resized to {new_width}x{new_height} due to high blue content.")
            else:
                print(f"Image at {image_path} not resized as its current size is already optimal.")
    else:
        print(f"Image at {image_path} is not predominantly blue, no resizing performed.")
```
