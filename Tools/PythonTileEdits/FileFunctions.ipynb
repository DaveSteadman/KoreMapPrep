```python
import os

def list_cs_files(root_dir):
    cs_files = []
    root_len = len(root_dir)
    for foldername, subfolders, filenames in os.walk(root_dir):
        for filename in filenames:
            if filename.endswith('.cs'):
                relative_path = os.path.join(foldername[root_len:].lstrip(os.path.sep), filename)
                cs_files.append(relative_path)
    return cs_files

# Replace 'your_directory_path' with the path to the directory you want to traverse
directory_path = 'E:/Data/03-CodeProjects/21-UnityProjects/Globe2/Assets/Code'
cs_files = list_cs_files(directory_path)

for file_path in cs_files:
    print(file_path)
```

```python
import os

def list_directories(root_dir):
    directories = []
    for foldername, subfolders, filenames in os.walk(root_dir):
        directories.append(foldername)
    return directories

# Replace 'your_directory_path' with the path to the directory you want to traverse
directory_path = 'your_directory_path'
directories = list_directories(directory_path)

for directory in directories:
    print(directory)
```
