```python


from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images in Jupyter Notebook
def display_image(image, title=''):
    plt.figure(figsize=(8, 8))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load the low-resolution satellite image
image_path = 'your_image_file.jpg'  # Replace with your image file path
image_path = 'C:/Util/Godot/Globe4-Maps/MapLib/Lvl3_0p2x0p2/BG_DA_CD/Sat_BG_DA_CD_DB.png'

low_res_image = cv2.imread(image_path)

# Display the original low-resolution image
display_image(low_res_image, title='Original Low-Resolution Image')

# Step 1: Upscale the image using different interpolation methods
# Method: cv2.INTER_CUBIC (bicubic interpolation)
upscale_factor = 2  # You can change this to the desired scaling factor
upscaled_image = cv2.resize(low_res_image, (low_res_image.shape[1] * upscale_factor, low_res_image.shape[0] * upscale_factor), interpolation=cv2.INTER_CUBIC)

# Display the upscaled image
display_image(upscaled_image, title='Upscaled Image (Bicubic Interpolation)')

# Step 2: Sharpen the upscaled image
# Define a kernel for sharpening
sharpening_kernel = np.array([[-1, -1, -1],
                              [-1,  9, -1],
                              [-1, -1, -1]])

# Apply the sharpening filter
sharpened_image = cv2.filter2D(upscaled_image, -1, sharpening_kernel)

# Display the sharpened image
display_image(sharpened_image, title='Sharpened Image')

# Optional: Save the processed image
output_path = 'sharpened_image.jpg'
output_path = 'C:/Util/Godot/Globe4-Maps/MapLib/Lvl3_0p2x0p2/BG_DA_CD/Sat_BG_DA_CD_DB2.png'

cv2.imwrite(output_path, sharpened_image)
print(f"Processed image saved as {output_path}")
```
